```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/Tata Dataset.csv')  #'/content/Tata Dataset.csv' the file path
print(data.head())

# Basic statistics of the dataset
print(data.describe())

# Plotting stock prices over time
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Analyzing volume
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Volume'], color='orange')
plt.title('Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```
